```python
from bs4 import BeautifulSoup
import json
```

```python
with open('Basic Materials.htm') as f:
    lines=f.readlines()
    page1=BeautifulSoup(lines,'lxml')
    summary=page1.find('Sector')
    print(summary)
```

```python
with open('Basic Materials.htm') as f:
    lines=f.readlines()
    Materials_json = json.dumps(lines)
    page1=BeautifulSoup(Materials_json,'lxml')
    print(page1.find('div'))
```
