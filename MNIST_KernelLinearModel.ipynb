```python
import numpy as np
import tensorflow as tf

```

```python
def get_input_fn(dataset, batch_size, capacity=10000, min_after_dequeue=3000):
    def  _input_fn():
        images_batch,labels_batch = tf.train.shuffle_batch(
            tensors=[dataset.images, dataset.labels.astype(np.int32)],batch_size=batch_size,capacity=capacity,
            min_after_dequeue=min_after_dequeue, enqueue_many=True)
        features_map={'images':images_batch}
        return features_map, labels_batch
    return _input_fn
```

```python
data=tf.contrib.learn.datasets.mnist.load_mnist()
```

```python

            

train_input_fn = get_input_fn(dataset=data.train, batch_size=256)
test_input_fn = get_input_fn(dataset=data.test, batch_size=5000)            
                                                          
```

```python
image_column = tf.contrib.layers.real_valued_column('images', dimension=784)
```

```python
optimizer= tf.train.FtrlOptimizer(learning_rate=1.0, l2_regularization_strength=1.0)
```

```python

estimator = tf.contrib.learn.LinearClassifier(feature_columns=[image_column],n_classes=10, optimizer=optimizer)
```

```python
estimator.fit(input_fn=train_input_fn,steps=2000)
```

```python
# Creates a _RealValuedColumn for dense numeric data.
# specify features for the estimator



# train 

# evaluate
results=estimator.evaluate(input_fn=test_input_fn, steps=1)
print(results)
```
